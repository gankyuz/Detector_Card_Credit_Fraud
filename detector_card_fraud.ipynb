Padronização de dados da coluna "Amount" e "Time".

```python
from sklearn.preprocessing import RobustScaler
new_df = df.copy()
new_df['Amount'] = RobustScaler().fit_transform(new_df['Amount'].to_numpy().reshape(-1,1))
new_df['Amount'].hist()
time = new_df['Time']
new_df['Time'] = (time - time.min()) / (time.max() - time.min())
new_df
```

Transformação de dados para um intervalo muito menor(entre 0 e 25) e diminuiu significativamente o desvio padrão da coluna.

Reescalonamento de todos os dados da coluna "time" para que passem a ser valores limpos contidos estritamente no intervalo entre 0 e 1.


### Divisão de dados em conjuntos de treino, teste e validação

```python
new_df = new_df.sample(frac=1, random_state=1)
new_df
```

```python
train, test, val = new_df[:240000], new_df[240000:262000], new_df[262000:]
train["Class"].value_counts(), test["Class"].value_counts(), val["Class"].value_counts()
```

```python
train_np, test_np, val_np = train.to_numpy(), test.to_numpy(), val.to_numpy()
train_np.shape, test_np.shape, val_np.shape
```

Separar features e target

```python
x_train, y_train = train_np[:, :-1], train_np[:,-1]
x_test, y_test = test_np[:, :-1], test_np[:,-1]
x_val, y_val = val_np[:, :-1], val_np[:,-1]

x_train.shape, y_train.shape, x_test.shape, y_test.shape, x_val.shape, y_val.shape
```

### Modelo de regressão de logística

Treinar um modelo de regressão de logística com dados de treino para calcular a acurácia do modelo nos mesmos dados.

```python
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression()
logistic_model.fit(x_train, y_train)
logistic_model.score(x_train, y_train)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_val, logistic_model.predict(x_val), target_names=["Não fraudulento", "Fraude"]))
```

Os dados estão desbalanceados, contudo, a acurácia é ilusória. Nesse caso, o algoritmo aprendeu a priorizar a classe majoritária, que é não fraudulento. Isso gera neglegenciamento a classe minoritária, que é fraude.

Conclusão: este modelo não serve para detectar fraudes com dados tão desbalanceados.

### Rede neural rasa

Criar uma rede neural simples para classificação binária, com camada oculta pequena, normalização e saída sigmoide.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, Dense, BatchNormalization
from tensorflow.keras.callbacks import ModelCheckpoint

shallow_nn = Sequential()
shallow_nn.add(InputLayer((x_train.shape[1],)))
shallow_nn.add(Dense(2, 'relu'))
shallow_nn.add(BatchNormalization())
shallow_nn.add(Dense(1, 'sigmoid'))

# salvar melhor modelo
checkpoint = ModelCheckpoint('shallow_nn.keras', save_best_only=True)
shallow_nn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
shallow_nn.summary()
```

```python
shallow_nn.fit(x_train, y_train, validation_data=(x_val, y_val), epochs=5, callbacks=checkpoint)
```

```python
def neural_net_predictions(model, x):
    return(shallow_nn.predict(x).flatten() > 0.5).astype(int)

neural_net_predictions(shallow_nn, x_val)
```

```python
print(classification_report(y_val, neural_net_predictions(shallow_nn, x_val), target_names=['Não fraudulento', 'Fraude']))
```

Este modelo apresenta um recall maior para a classe fraude, podendo identificar mais casos de fraude, porém com uma precisão menor.

O caso da acurácia é o mesmo da regressão logística, parecem perfeitas por causa do desbalanceamento.

- Se a prioridade é não deixar fraudes passarem, a rede neural é mais adequada.
- Se a prioridade é minimizar falsos alarmes, a regressão logística pode ser escolhida.

Conclusão: ambos os modelos possuem uma acurácia semelhante, a rede neural apresenta maior sensibilidade para identificar fraudes, enquanto a regressão logistíca privilegia a precisão.



### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(max_depth=2, n_jobs=-1)
rf.fit(x_train, y_train)
print(classification_report(y_val, rf.predict(x_val), target_names=['Não fraudulento', 'Fraude']))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gbc= GradientBoostingClassifier(n_estimators=50, learning_rate=1.0, max_depth=1, random_state=0)
gbc.fit(x_train, y_train)
print(classification_report(y_val, gbc.predict(x_val), target_names=['Não fraudulento', 'Fraude']))
```

```python
from sklearn.svm import LinearSVC
svc = LinearSVC(class_weight='balanced')
svc.fit(x_train, y_train)
print(classification_report(y_val, svc.predict(x_val), target_names=['Não fraudulento', 'Fraude']))
```
